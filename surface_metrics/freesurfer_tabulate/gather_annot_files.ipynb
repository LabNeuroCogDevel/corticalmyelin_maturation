## Volumetric atlases to the surface

These atlases need to be transformed into surface space before we can get them into annot format. We're going to download the region names and nifti files from neuroparc and then transform them into giftis using neuromaps.

We'll be mapping:

 * AAL
 * HOCPATh25
 * Juelich
 * Slab


```python
%%bash

mkdir -p neuroparc
cd neuroparc

# AAL
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/aal_space-MNI152NLin6_res-1x1x1.nii.gz
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/aal_space-MNI152NLin6_res-1x1x1_updated.json

# HOCPATh25
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/HarvardOxfordcort-maxprob-thr25_space-MNI152NLin6_res-1x1x1.nii.gz
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/HarvardOxfordcort-maxprob-thr25_space-MNI152NLin6_res-1x1x1_updated.json

# Juelich
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/Juelichgmthr252mmEickhoff2005.nii.gz
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/Juelichgmthr252mmEickhoff2005.json

# Slab
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/slab907_space-MNI152NLin6_res-1x1x1.nii.gz
wget https://github.com/neurodata/neuroparc/raw/v1.0/atlases/label/Human/slab907_space-MNI152NLin6_res-1x1x1_updated.json

3dLRflip -LR -prefix neuroparc_aal_lrflip.nii.gz aal_space-MNI152NLin6_res-1x1x1.nii.gz
```

```python
import nibabel as nb
import numpy as np
import json
from neuromaps import transforms
import re


atlases_to_map = [
    ("AAL",
     "neuroparc/neuroparc_aal_lrflip.nii.gz",
     "neuroparc/aal_space-MNI152NLin6_res-1x1x1_updated.json"),
    ("HOCPATh25",
     "neuroparc/HarvardOxfordcort-maxprob-thr25_space-MNI152NLin6_res-1x1x1.nii.gz",
     "neuroparc/HarvardOxfordcort-maxprob-thr25_space-MNI152NLin6_res-1x1x1_updated.json"),
    ("Juelich",
     "neuroparc/Juelichgmthr252mmEickhoff2005.nii.gz",
     "neuroparc/Juelichgmthr252mmEickhoff2005.json"),
    ("Slab",
     "neuroparc/slab907_space-MNI152NLin6_res-1x1x1.nii.gz",
     "neuroparc/slab907_space-MNI152NLin6_res-1x1x1_updated.json"
    ),
    ("CC200",
     "cpac_templates/CC200.nii.gz",
     None
    ),
    ("CC400",
     "cpac_templates/CC400.nii.gz",
     None
    )
]


def remove_non_alphabetic(input_string):
    # Use regular expression to remove non-alphabetic characters
    clean_string = re.sub(r'[^a-zA-Z_0-9]', '', input_string.replace(" ", "_"))
    return clean_string


def _create_colors(n_colors):
    color_set = set([(0,0,0, 0)])
    while len(color_set) < n_colors:
        new_color = tuple((np.random.rand(3)*155).astype(np.int32)) + (0,)
        color_set.add(new_color)
    color_mat = np.array(sorted(color_set))
    assert color_mat.shape[0] == n_colors
    return color_mat


def fake_neuroparc_from_nifti(nifti_file):
    img = nb.load(nifti_file)
    unique_labels = np.unique(img.get_fdata().astype(np.int32))
    return {str(label):{} for label in unique_labels}


def fill_missing_parc(spec):
    
    maxval = max(map(int, spec.keys()))
    for key in range(maxval):
        strkey = str(key)
        if strkey not in spec:
            spec[strkey] = {"label": "Unknown"}
    return spec

            
def ctab_from_neuroparc_json(neuroparc_json_file=None, atlas_nifti_file=None):

    if neuroparc_json_file is not None:
        with open(neuroparc_json_file, "r") as jsonf:
            initial_specs = json.load(jsonf)
            if "MetaData" in initial_specs:
                del initial_specs["MetaData"]
    else:
        initial_specs = fake_neuroparc_from_nifti(atlas_nifti_file)

    parc_specs = fill_missing_parc(initial_specs)
    # Get a mapping from ints to label names from neuroparc
    int_to_label_map = {}
    for key, value in parc_specs.items():
        if not key.isnumeric():
            continue
        key = int(key)
        label = value.get("label") or "region%05d" % key
        int_to_label_map[key] = label
    int_to_label_map[0] = "Unknown"
    labels = [remove_non_alphabetic(int_to_label_map[key]) for key in 
              sorted(int_to_label_map.keys())]
    colors = _create_colors(len(labels))
    return colors, labels


def create_annots(nifti_file, output_stem, json_file=None):
    
    lh_gii, rh_gii = transforms.mni152_to_fsaverage(nifti_file, 
                                            fsavg_density="164k", 
                                            method="nearest")

    colors, names = ctab_from_neuroparc_json(neuroparc_json_file=json_file,
                                             atlas_nifti_file=nifti_file)
    nb.freesurfer.write_annot("annots/lh." + output_stem + ".annot",
                              labels=lh_gii.agg_data().astype(np.int32),
                              ctab=colors,
                              names=names,
                              fill_ctab=True)
    nb.freesurfer.write_annot("annots/rh." + output_stem + ".annot",
                          labels=rh_gii.agg_data().astype(np.int32),
                          ctab=colors,
                          names=names,
                          fill_ctab=True)

for atlasname, nifti_file, json_file in atlases_to_map:
    print(atlasname)
    create_annots(nifti_file, atlasname, json_file)
```
